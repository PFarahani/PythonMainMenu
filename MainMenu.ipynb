```python
def FeatureMenu(status):
    
    #########################################
    ############# Functions #################
    
    def showFeatureMenu(status):
        os.system('cls')

        # Mapping status to a string
        s = ["Not Active", "Active"]
        d = dict([(x,y) for x,y in enumerate(set(s))])
        status_s = [d[x] for x in status]

        print("""
============================================================
                        Features
============================================================
1: Save all generated tables                     {}""".format(status_s[0]),"""

2: Save the final output as a pivot table        {}""".format(status_s[1]),"""


0: Return to main menu                                     
============================================================""")
        print("""
Please enter your choice:
        """)
    
    def feature1(status):
        if status==0:
            return 1
        elif status==1:
            return 0
        else:
            print("Invalid value!")
            pass
    def feature2(status):
        if status==0:
            return 1
        elif status==1:
            return 0
        else:
            print("Invalid value!")
            pass  
    
    #########################################
    #########################################
    
    # Show the initial state of the FeaturesMenu
    showFeatureMenu(status)
    while True:
        try:
            choice = int(input())
            if choice==1:
                status[0] = feature1(status[0])
                os.system('cls')
                showFeatureMenu(status)
                continue
            elif choice==2:
                status[1] = feature2(status[1])
                os.system('cls')
                showFeatureMenu(status)
                continue
            elif choice==0:
                os.system('cls')
                return status
                break
            if not 0<=choice<=2:
                print("Invalid input! Try again.")
                continue

        except ValueError:
            print("Unrecognized command! Try again.")
            continue
```

```python
def MainMenu():
    
    #########################################
    ############# Functions #################

    def showMainMenu():    
        print("""
============================================================
                         Main Menu                         
============================================================
1: Main process

2: Features


0: Exit
============================================================
    
    """)
        print("Please Enter your choice:")
    
    
    #########################################
    #########################################
    
    # Show the initial state of the FeaturesMenu
    showMainMenu()
    
    # Optional features initial status
    status = [0,0]
    while True:
        try:
            choice = int(input())
            if choice==1:
                os.system('cls')
            # This in fact breaks this loop and takes you to the main process
                break
            elif choice==2:
                os.system('cls')
                status = FeatureMenu(status)
                os.system('cls')
                showMainMenu()
                continue
            elif choice==0:
                sys.exit()
            elif not 0<=choice<=2:
                print("Invalid input! Try again.")
                continue
        except ValueError:
            print("Unrecognized command! Try again.")
            continue                
                
    return status
```
